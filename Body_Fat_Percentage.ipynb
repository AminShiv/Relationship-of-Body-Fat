One of the most crucial aspects of maintaining overall health and well-being revolves around comprehending and effectively managing body fat. The body fat percentage, denoting the proportion of fat to total body mass, exerts a profound influence on our aesthetic appearance, fitness levels, and susceptibility to various illnesses. Establishing and sustaining a healthy body fat percentage is particularly advantageous due to the strong correlation between excess body fat and an elevated risk of numerous health problems. Notably, cardiovascular diseases, diabetes, and various metabolic disorders are among the potential repercussions of being overweight.

Conversely, it is essential to recognize that maintaining an excessively low body fat percentage can also give rise to health concerns, disrupting essential bodily processes. Knowing the right balance between overweight and underweight is the key to promoting optimal well-being. To gain a comprehensive understanding of fat distribution throughout the body, measuring body fat percentage becomes a valuable tool. This information not only informs individuals about their body composition but also provides them important information for dieting plans and fitness plans.

Various methods exist for obtaining accurate body fat percentage data. Techniques such as skinfold calipers, bioelectrical impedance analysis, and dual-energy X-ray absorptiometry (DEXA) are recognized as among the most precise ways to calculate and determine body fat. These methods offer insights into the distribution of fat, providing individuals in making informed decisions about their health and fitness.

# **<span style = "color: lightblue"> Methods </span>** 


### Data Collection

I have got this data set from Kaggle.com

```python
from IPython.display import display
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


body_fat = pd.read_csv("BodyFat - Extended.csv")
print(body_fat.shape)
display(body_fat)
```

# **<span style = "color: lightblue"> Data Cleaning </span>** 

For the data cleaning portion, all of my data were full and nothing really needed to be change. I just needed to put the decimal places to 1 decimal place so the data looks more organized and I also changed the Original column from Y and N to 1's and 0's to make it simplier to read.




```python
pd.set_option('display.precision', 1)
body_fat = body_fat.replace({'Original': {'Y': 1, 'N': 0}})
display(body_fat)
```

I used the describe built-in function to look at each column's mean because I will be using this data set later on in my investigation.

```python
body_fat.describe()
```

# **<span style = "color: lightblue"> Analysis </span>** 

### **<span style = "color: yellow"> Age Count </span>** 

First I made a graph that looked at the count of each age group their is in the data set. From the graph, most of the ages fall in the age range of 19-23, mostly in the age of 20. 

```python
body_fat['Age'].value_counts().sort_index().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Age Count')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.gcf().set_size_inches(12, 6)
plt.show()
```

For the below code, I have just diplayed what the mean, maximun and minimum values were for the age.

```python
body_fat['Age'] = pd.to_numeric(body_fat['Age'], errors='coerce')
age_bins = list(range(0, 100, 10))
age_labels = [f'{start}-{start + 9}' for start in age_bins[:-1]]
body_fat['AgeGroup'] = pd.cut(body_fat['Age'], bins=age_bins, labels=age_labels, right=False)
grouped_data = body_fat.groupby('AgeGroup')['BodyFat'].agg(['mean', 'max', 'min'])
print(grouped_data)
```

### **<span style = "color: yellow"> Gender Count </span>** 


I made this graph to show the count of each gender just to get a just of how many male and female are contained in my data set.

```python
body_fat['Sex'].value_counts().sort_index().plot(kind='bar', color=['Blue', 'Red'])
plt.title('Male and Female Count')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

For the below code, I have just diplayed what the mean, maximun and minimum values were for the gender.

```python
grouped_data = body_fat.groupby('Sex')['BodyFat'].agg(['mean', 'max', 'min'])
print(grouped_data)
```

### **<span style = "color: yellow"> Body Fat Distribution for Male and Female </span>** 

I have made a graph to look at the relationship to see what is the body fat percentage distribution between male and female. From the graph, we can see that male have a lower body fat percentage compaired to females.

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='BodyFat', data=body_fat, palette='pastel')
plt.title('Distribution of Body Fat by Sex')
plt.xlabel('Sex')
plt.ylabel('Body Fat Percentage')
plt.show()
```

### **<span style = "color: yellow"> Body Fat Distribution for Ages </span>** 

Looking at the graph below for Ages vs Body Fat Percentages, we can see that in your early ages of life and later ages, you have more body fat percentages then ages in between. Their are some ages which doesn't count towards that statement like 46, 58, and 68 but the overall pattern can be seen. Another observation that can be made is that the average body fat percentage is between 15% - 25%. To be more accurate about this observation, I have made a graph below this graph to show where most of the body fat percentage is bound.

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Age', y='BodyFat', data=body_fat, palette='pastel')
plt.xticks(rotation=45, ha='left')
plt.gcf().set_size_inches(20, 10)
plt.title('Distribution of Body Fat by Sex')
plt.xlabel('Age')
plt.ylabel('Body Fat Percentage')
plt.show()
```

This graph proves that most of the bodyfat was around 15-20 percentage.

```python
sns.histplot(body_fat['BodyFat'], bins=20, kde=True)
plt.title('Body Fat Distribution')
plt.xlabel('Body Fat')
plt.ylabel('Frequency')
plt.show()
```

### **<span style = "color: yellow"> Grouped Bar Plot by Gender for Various Body Fat Percentages </span>** 

Looking at my final graph for my analysis, I have made a grouped bar graph which shows each body part of for male and female and it shows the comparison between each body part. From this graph, we can see that the average body fat percentage of a male is more then female. Some of the body parts have a big difference like Abdomen, Chest, and Thigh. The other have a smaller difference like Ankle, Wrist, Bicep and Forearm. 

```python
group = body_fat.groupby('Sex')[['Neck', 'Chest', 'Abdomen', 'Hip', 'Thigh', 'Knee', 'Ankle', 'Biceps', 'Forearm', 'Wrist']].mean()
plt.figure(figsize=(20,10))
group.plot(kind = 'bar', colormap='tab20')
plt.title('Grouped Bar Plot by Gender for Average Body Fat Percentages')
plt.xlabel('Sex')
plt.ylabel('Average Body Fat Percentages')
plt.xticks(rotation=0, ha='left')
plt.legend(title='Measurement', bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```
